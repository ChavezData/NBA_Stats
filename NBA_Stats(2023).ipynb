```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_stats = pd.read_csv('2023_nba_player_stats.csv')
display(df_stats.head(10))
```

# NBA Player Statistics Analysis

This notebook analyzes the 2023 NBA player statistics dataset to explore various aspects of player performance. We will be looking at key metrics such as points, rebounds, assists, and more to gain insights into the top performers and overall player distributions.

```python
top_5_scorers = df_stats.sort_values(by='PTS', ascending=False).head(5)
display(top_5_scorers[['PName', 'Team', 'PTS']])
```

```python
display(top_5_scorers[['PName', 'Team', 'AST']])
```

```python
display(top_5_scorers[['PName', 'Team', 'REB']])
```

```python
display(top_5_scorers[['PName', 'Team']])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_stats['PTS'], bins=30, kde=True)
plt.title('Distribution of Player Points')
plt.xlabel('Points')
plt.ylabel('Number of Players')
plt.show()
```

The histogram shows the distribution of points for all players in the dataset. We can see that the distribution is skewed, with a large number of players scoring fewer points and a smaller number of players scoring significantly more.

```python
display(top_5_scorers[['PName', 'Age']])
```

```python
top_5_scorers_melted = top_5_scorers.melt(id_vars=['PName', 'Team'], value_vars=['PTS', 'REB', 'AST'], var_name='Statistic', value_name='Value')

plt.figure(figsize=(14, 8))
sns.barplot(x='PName', y='Value', hue='Statistic', data=top_5_scorers_melted)
plt.title('Comparison of Points, Rebounds, and Assists for Top 5 Scorers')
plt.xlabel('Player Name')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Statistic')
plt.tight_layout()
plt.show()
```

By looking at the height of the bars for each player, you can easily compare their performance in terms of scoring, rebounding, and assisting. For example, you can see who had the most points, who had the most rebounds, and who had the most assists among these top 5 players.
